# Simulating random galaxies

In this notebook, I will use galsim to simulate random galaxies.

## Downloading libraries

```python
# Load modules/libraries
import copy
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
from astropy.io import fits
import galsim
import sys
import os
import math
import logging
import time
```

We will create elliptical galaxies, spiral galaxies, ring galaxies and mergers of two galaxies. We will do 75:100:100:125.

## Creating one elliptical galaxy


```python
##### parameter values
#we keep the flux the same for now
bulge_frac = round(np.random.uniform(0.1,0.9),1) #0.5
bulge_hlr = round(np.random.uniform(1,20),1) #10
angle = round(np.random.uniform(0,180),1) #30
disk_hlr = round(np.random.uniform(1,20),1) #2.5
if (disk_hlr+bulge_hlr)>20:
    disk_hlr =round(np.random.uniform(1,10),1)
    bulge_hlr=round(np.random.uniform(1,10),1)
ellipticity = round(np.random.uniform(0.1,0.9),1)#0.5

# Increase the maximum FFT size allowed by GalSim
galsim.GSParams.maximum_fft_size = 18000

# Set up the parameters for the galaxy
galaxy_params = {
    'flux': 1e5,                # Total flux of the galaxy
    'bulge_frac': bulge_frac,          # Fraction of light in the bulge component
    'bulge_hlr': bulge_hlr,           # Half-light radius of the bulge component (in arcseconds)
    'disk_hlr': disk_hlr,            # Half-light radius of the disk component (in arcseconds)
    'ellipticity': ellipticity,         # Ellipticity of the galaxy
    'angle': angle               # Rotation angle of the galaxy (in degrees)
}

# Create a galaxy using a bulge-disk model
bulge = galsim.DeVaucouleurs(half_light_radius=galaxy_params['bulge_hlr'])
disk = galsim.Exponential(half_light_radius=galaxy_params['disk_hlr'])
galaxy = galaxy_params['bulge_frac'] * bulge + (1 - galaxy_params['bulge_frac']) * disk

# Apply additional transformations
galaxy = galaxy.shear(e=galaxy_params['ellipticity'], beta=galaxy_params['angle']*galsim.degrees)

# Create an image of the galaxy
image_size = 64  # Size of the square image in pixels
pixel_scale = 0.2  # Size of each pixel in arcseconds
image = galsim.ImageF(image_size, image_size, scale=pixel_scale)
galaxy.drawImage(image)

# Display the generated galaxy image
plt.imshow(image.array,  origin='lower', extent=[-image_size/2, image_size/2, -image_size/2, image_size/2])
plt.colorbar(label='Flux')
plt.title('Using variables: {}'.format(galaxy_params) )
plt.show()
```

## Simulate ring galaxies 

```python
import galsim
import numpy as np
import matplotlib.pyplot as plt

def generate_spiral_galaxy(image_size, pixel_scale, random_seed=None):
    if random_seed is not None:
        rng = galsim.BaseDeviate(random_seed)
    else:
        rng = None

    bulge_flux = 1.e5
    bulge_r0 = 0.5  # Scale radius of the bulge

    spiral_arm_amplitude = 5.e4
    spiral_arm_wavelength = 10.0  # Wavelength of the spiral arms
    spiral_arm_pitch_angle = 30.0  # Pitch angle of the spiral arms (in degrees)

    # Bulge component
    bulge = galsim.Exponential(flux=bulge_flux, scale_radius=bulge_r0)

    # Create a grid of coordinates
    x, y = np.meshgrid(np.arange(image_size) - image_size / 2, np.arange(image_size) - image_size / 2)

    # Spiral arms parametric equations
    r = np.sqrt(x**2 + y**2)
    theta = np.arctan2(y, x)
    spiral_arms = np.cos(np.radians(spiral_arm_pitch_angle) * r)

    # Modulate the amplitude of the spiral arms with a Gaussian
    gaussian_amplitude = np.exp(-r**2 / (2 * spiral_arm_wavelength**2))
    spiral_arms *= gaussian_amplitude * spiral_arm_amplitude

    # Combine components
    gal = bulge + galsim.InterpolatedImage(galsim.Image(spiral_arms, scale=pixel_scale))

    # Convolve with a PSF (optional)
    psf_beta = 5
    psf_re = 1.0
    psf = galsim.Moffat(beta=psf_beta, flux=1., half_light_radius=psf_re)
    final = galsim.Convolve([gal, psf])

    # Draw the image with a particular pixel scale.
    image = final.drawImage(scale=pixel_scale)
    return image

def visualize_spiral_galaxy(image_size, pixel_scale, random_seed=None):
    galaxy_image = generate_spiral_galaxy(image_size, pixel_scale, random_seed=random_seed)

    plt.imshow(galaxy_image.array, origin='lower', cmap='viridis', extent=[0, image_size, 0, image_size])
    plt.show()

# Example usage
image_size = 100
pixel_scale = 0.2
random_seed = 42
visualize_spiral_galaxy(image_size, pixel_scale, random_seed=random_seed)
```

## Simulating 1 ring galaxy

```python
gal_flux = 1.e5       # Total flux of the galaxy
gal_hlr = round(np.random.uniform(2,5),2)        # Half-light radius of the galaxy
gal_e = round(np.random.uniform(0, 1),2) #0.5          # Ellipticity of the galaxy, [0,1]
gal_beta = round(np.random.uniform(0, 180),2)#45      # Position angle of the galaxy in degrees
ring_frac = 0.2     # Fraction of the total flux in the ring
ring_hlr = 0.3       # Half-light radius of the ring
ring_hlr_outer = round(np.random.uniform(0.1, 5),2)#2.5  # Outer half-light radius of the ring
ring_hlr_inner = round(np.random.uniform(1,5),2) #2  # Inner half-light radius of the ring

size = (64,64)

# Create the Gaussian galaxy
gal = galsim.Gaussian(flux=gal_flux, half_light_radius=gal_hlr)

# Apply ellipticity to the galaxy using Shear
shear = galsim.Shear(e=gal_e, beta=gal_beta * galsim.degrees)
gal = gal.shear(shear)

# Create the outer Gaussian for the ring
ring_outer = galsim.Gaussian(flux=ring_frac * gal_flux, half_light_radius=ring_hlr_outer)

# Create the inner Gaussian for the hole in the ring
ring_inner = galsim.Gaussian(flux=ring_frac * gal_flux, half_light_radius=ring_hlr_inner)

# Subtract the inner Gaussian from the outer Gaussian to create the ring
ring = ring_outer - ring_inner

# Combine galaxy and ring to create the final ring galaxy
final_galaxy = gal + ring

image = final_galaxy.drawImage(scale=0.2)
plt.imshow(image.array, cmap='gray',extent=[0, size[0], 0, size[1]])
plt.title(f'Using variables {gal_hlr}, {gal_e}, {gal_beta}, {ring_frac}, {ring_hlr}')

plt.show()
```

## Simulating 100 ring galaxies

```python
#basic starting parameters
gal_flux = 1.e5       # Total flux of the galaxy
ring_frac = 0.2     # Fraction of the total flux in the ring
ring_hlr = 0.3       # Half-light radius of the ring

for i in range(50):
    
    #randomize the parameters
    gal_hlr = round(np.random.uniform(2,5),2)        # Half-light radius of the galaxy
    gal_e = round(np.random.uniform(0, 1),2) #0.5          # Ellipticity of the galaxy, [0,1]
    ring_hlr_outer = round(np.random.uniform(0.1, 5),2)#2.5  # Outer half-light radius of the ring
    ring_hlr_inner = round(np.random.uniform(1,5),2) #2  # Inner half-light radius of the ring
    gal_beta = round(np.random.uniform(0, 180),2)#45      # Position angle of the galaxy in degrees
    
    # Create the Gaussian galaxy
    gal = galsim.Gaussian(flux=gal_flux, half_light_radius=gal_hlr)

    # Apply ellipticity to the galaxy using Shear
    shear = galsim.Shear(e=gal_e, beta=gal_beta * galsim.degrees)
    gal = gal.shear(shear)

    # Create the outer Gaussian for the ring
    ring_outer = galsim.Gaussian(flux=ring_frac * gal_flux, half_light_radius=ring_hlr_outer)

    # Create the inner Gaussian for the hole in the ring
    ring_inner = galsim.Gaussian(flux=ring_frac * gal_flux, half_light_radius=ring_hlr_inner)

    # Subtract the inner Gaussian from the outer Gaussian to create the ring
    ring = ring_outer - ring_inner

    # Combine galaxy and ring to create the final ring galaxy
    final_galaxy = gal + ring

    image = final_galaxy.drawImage(scale=0.2)
    plt.imshow(image.array, cmap='gray',extent=[0, size[0], 0, size[1]])
    
    filename = f'random_sim_gal{i}_ring.png'
    save_path = f'random_sim_gal/{filename}'
    print(save_path)
    plt.savefig(save_path)
    plt.show()

    png_image = Image.open(save_path)
    # Convert the PNG image to a NumPy array
    data = np.array(png_image)

# Create a FITS header
    header = fits.Header()
    header['COMMENT'] = 'Converted from PNG to FITS'

# Create a FITS HDU (Header/Data Unit)
    hdu = fits.PrimaryHDU(data, header=header)

# Specify the path for saving the FITS file
    filename = f'ring{i+100}.fits'
    fits_path = f'random_sim_gal/{filename}'

# Write the FITS file
    hdu.writeto(fits_path, overwrite=True)

    if os.path.exists(save_path):
        # Delete the file
        os.remove(save_path)
```
